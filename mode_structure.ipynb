```python
import matplotlib.pyplot as plt

# Data for plotting the model horizontally with adjusted spacing and fontsize
layers_horizontal_adjusted = [
    ("Input Shape", (0, 5)),
    ("Conv1D\n(filters=64,\nkernel_size=3)", (2.5, 5)),
    ("MaxPooling1D\n(pool_size=2)", (5, 5)),
    ("LSTM\n(units=64,\nreturn_sequences=True)", (7.5, 5)),
    ("LSTM\n(units=64)", (10, 5)),
    ("Dense\n(units=64)", (12.5, 5)),
    ("Dropout\n(rate=0.3)", (15, 5)),
    ("Dense\n(units=32)", (17.5, 5)),
    ("Dropout\n(rate=0.5)", (20, 5)),
    ("Dense\n(units=1,\nactivation=sigmoid)", (22.5, 5)),
]

# Create the figure and axis for horizontal layout
fig, ax = plt.subplots(figsize=(24, 8))

# Plot each layer as a box with text
for (label, (x, y)) in layers_horizontal_adjusted:
    ax.text(x, y, label, ha='center', va='center', fontsize=10,
            bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightblue'))

# Draw arrows between the layers
for i in range(len(layers_horizontal_adjusted) - 1):
    (x1, y1) = layers_horizontal_adjusted[i][1]
    (x2, y2) = layers_horizontal_adjusted[i + 1][1]
    ax.annotate("", xy=(x2-0.5, y2), xytext=(x1+0.5, y1), 
                arrowprops=dict(arrowstyle="->", color='black'))

# Adjust plot limits and hide axes
ax.set_xlim(-1, 24)
ax.set_ylim(4, 6)
ax.axis('off')

# Set the title
plt.title('Model Structure')
plt.show()
```
